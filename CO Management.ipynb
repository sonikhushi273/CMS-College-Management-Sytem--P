```python
def menu():
        print("                 Our Menu                        ")
        print("S.No.           Items Name                Amount")
        print("================================================")
        print("1.               Patties                    ₹20")
        print("2.               Samosa                     ₹15")
        print("3.               Maggie                     ₹30")
        print("4.               Pasta                      ₹35")
        print("5.               Sandwich                   ₹25")
        print("6.               Burger                     ₹30")
        print("7.               Coffee                     ₹20")
        print("8.               Cold Drink                 ₹20")

def order_now():
    order = {}
    while True:
        i = input("Enter Item Number : ")
        if i == '0':
            break
        elif i in ['1','2','3','4','5','6','7','8']:
            quan = int(input("Enter Quantity: "))
            order[i] = quan
            
        else:
            print("Entered Item is not in Our Menu")
    return order

def calculate_total(order):
    total = 0
    for i, quan in order.items():
        if i == '1':
            total += 20 * quan
        elif i == '2':
            total += 15 * quan
        elif i == '3':
            total += 30 * quan
        elif i == '4':
            total += 35 * quan
        elif i == '5':
            total += 25 * quan
        elif i == '6':
            total += 30 * quan
        elif i == '7':
            total += 20 * quan
        elif i == '8':
            total += 20 * quan
    return total

def canteen():
    print("Welcome to Canteen".center(100))
    menu()
    order = order_now()
    total = calculate_total(order)
    print("Your order is:")
    for i, quan in order.items():
        if i == '1':
            print("Patties x", quan)
        elif i == '2':
            print("Samosa x", quan)
        elif i == '3':
            print("Maggie x", quan)
        elif i == '4':
            print("Pasta x", quan)
        elif i == '5':
            print("Sandwich x", quan)
        elif i == '6':
            print("Burger x", quan)
        elif i == '7':
            print("Coffee x", quan)
        elif i == '8':
            print("Cold Drink x", quan)
    print("Total: ₹", total)
#canteen()
```
